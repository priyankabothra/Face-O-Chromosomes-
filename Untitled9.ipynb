```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
ImageDataGenerator(featurewise_center=False, samplewise_center=False, featurewise_std_normalization=False, samplewise_std_normalization=False, zca_whitening=False, zca_epsilon=1e-06, rotation_range=0, width_shift_range=0.0, height_shift_range=0.0, brightness_range=None, shear_range=0.0, zoom_range=0.0, channel_shift_range=0.0, fill_mode='nearest', cval=0.0, horizontal_flip=False, vertical_flip=False, rescale=None, preprocessing_function=None, data_format='channels_last', validation_split=0.0, dtype='float32')
```

```python
train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range= 15,
        brightness_range=(0.5,1.1),
        horizontal_flip=True,
        #width_shift_range=0.2,
        #zoom_range=0.3,
        #height_shift_range=0.2,
        fill_mode = "nearest",
        dtype= 'float32'
        )
 
test_datagen = ImageDataGenerator(rescale=1./255,
                                  horizontal_flip= True,
                                  #brightness_range= (0.2,1),
                                  dtype= 'float32'
                                  )
```

```python
train_generator = train_datagen.flow_from_directory(
    directory = "/gdrive/My Drive/Colab Notebooks/4people/people/Train",
    target_size=(140, 140),
    color_mode="grayscale",
    batch_size=4,
    class_mode="categorical",
    shuffle=True,
    seed=42
    #save_to_dir = "/gdrive/My Drive/4people/gen/train"
)
```

```python
valid_generator = test_datagen.flow_from_directory(
    directory="/gdrive/My Drive/Colab Notebooks/4people/people/Validate",
    target_size=(140, 140),
    color_mode="grayscale",
    batch_size=2,
    class_mode="categorical",
    shuffle=True,
    seed=42
    #save_to_dir = "/gdrive/My Drive/4people/gen/val"
)
```

```python
import numpy as np
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import LeakyReLU
import matplotlib.pyplot as plt
from tensorflow.keras.backend import clear_session
clear_session()
%matplotlib inline
```

```python
batch_size = 20
epochs = 50
num_classes = 6
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(140,140,1),padding='same'))
#model.add(LeakyReLU(alpha=0.3))
 
model.add(MaxPooling2D(pool_size=(2, 2),padding='same'))
 
model.add(Conv2D(64, (3, 3),activation='relu', padding='same'))
#model.add(LeakyReLU(alpha=0.1))
 
model.add(MaxPooling2D(pool_size=(2, 2),padding='same'))
 
model.add(Conv2D(128, (5, 5),activation='relu',padding='same'))
#model.add(LeakyReLU(alpha=0.1)) 
 
#model.add(MaxPooling2D((2, 2),padding='same'))
model.add(Conv2D(128, (5,5), activation= 'relu', padding='same'))
 
model.add(MaxPooling2D((2, 2),padding='same'))

model.add(Conv2D(256, (5,5), activation= 'relu', padding='same'))
model.add(MaxPooling2D((2, 2),padding='same'))
 
model.add(Flatten())
 
##model.add(Dense(512,activation='relu'))
model.add(Dense(256,activation='relu'))
#model.add(LeakyReLU(alpha=0.1)) 
 
model.add(Dropout(0.4))
 
#model.add(Dense(128,activation='relu'))
#model.add(LeakyReLU(alpha=0.1))
 
##model.add(Dense(32,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.2))
#model.add(LeakyReLU(alpha=0.1))
model.add(Dense(8,activation='relu'))
 
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.0001),metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ModelCheckpoint
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=30)
mc = ModelCheckpoint('/gdrive/My Drive/Colab Notebooks/finalmodel_r.h5', monitor='val_accuracy', mode='max', verbose=1, save_best_only=True)

print(train_generator.n, valid_generator.n)
print(train_generator.batch_size, valid_generator.batch_size)
```

```python
#from IPython.display import display
#from PIL import Imagee
STEP_SIZE_TRAIN=train_generator.n//train_generator.batch_size
STEP_SIZE_VALID=valid_generator.n//valid_generator.batch_size

history = model.fit(train_generator,
                    steps_per_epoch= STEP_SIZE_TRAIN,
                    validation_data=valid_generator,
                    validation_steps= STEP_SIZE_VALID,
                    epochs= 200,
                    callbacks=[mc]
                    
)
 
 
import matplotlib.pyplot as plt
# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
 
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
from tensorflow.keras.models import load_model
model = load_model('/gdrive/My Drive/Colab Notebooks/finalmodel_r.h5')
```

```python
#test_eval = model.evaluate_generator(generator=valid_generator)
model.evaluate(valid_generator, steps = STEP_SIZE_VALID)
```
